```python
from pyspark.sql.functions import *
```

```python
#silver
df_bronze=spark.read.format("delta").load("/Volumes/dbw_de_training/default/utkarshk/bronze/")

clean_df=(
    df_bronze.dropDuplicates().filter(col("Glucose") >0).filter(col("BloodPressure") >0).filter(col("BMI") >0)
)

silver_df=clean_df.select(
    col("Pregnancies").cast("int"),
    col("Glucose").cast("int"),
    col("BloodPressure").cast("int"),
    col("SkinThickness").cast("int"),
    col("Insulin").cast("int"),
    col("BMI").cast("double"),
    col("DiabetesPedigreeFunction").cast("double"),
    col("Age").cast("int"),
    col("Outcome").cast("int"),
    col("ingest_time")
)
```

```python
silver_df.write.format("delta").mode("overwrite").save("/Volumes/dbw_de_training/default/utkarshk/silver")
```
