```python
from pyspark.sql.functions import *
```

```python
gold_df=spark.read.format("delta").load("/Volumes/dbw_de_training/default/utkarshk/silver")
#gold
gold_df=gold_df.withColumn(
    "risk_category",
    when(col("Glucose")>140,"High Risk")
    .when(col("Glucose")>100,"Medium Risk").otherwise("Low Risk")
)
```

```python
gold_df.write.format("delta").mode("overwrite").save("/Volumes/dbw_de_training/default/utkarshk/gold/")
```
